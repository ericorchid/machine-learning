# 人工智能——时间线维度 from timeline

## 1950 图灵第一个定义智能，通过对话，无法分清对方是机器or人类的话，机器就是人工智能。（可理解、可测试）
图灵提出了现代计算机的核心，cpu，状态机，唯一问题在于无法决定何时停止
哲学（西方哲学入门、苏菲的世界）

## 1956 达特茅斯的定义AI：Every aspect of learning or any other feature of **intelligence** can be so precisely described that a machine can be made to **simulate** it.
能够被精确描述的智能，机器能够模拟它，就是人工智能

## 1969 感知机（前馈网络，dnn的爷爷爷）, 被明斯基Minsky证明是线性模型，无法解决**XOR**问题（现实中常见）将AI研究转向**Symbols** and **Logics**。符号运算和逻辑推理
该证明被认为说明函数拟合function fitting无法实现XOR，但当时没有意识到加激活函数，或者加维度的方式进行解决，导致没有发现dnn
